```python
import numpy as np # linear algebra
import pandas as pd
import os
```

```python
from fastai import *
from fastai.vision.all import *

trn_path = '/kaggle/input/140k-real-and-fake-faces/real_vs_fake/real-vs-fake/train'
files = get_image_files(trn_path)
```

```python
img = PILImage.create(files[20])
print(img.size)
img.to_thumb(128)
```

```python
dls = ImageDataLoaders.from_folder(trn_path, valid_pct=0.2, seed=42,
    item_tfms=None,
    batch_tfms=None)

dls.show_batch(max_n=12)
```

```python
learn = vision_learner(dls, 'resnet26d', metrics=accuracy, path='.').to_fp16()
```

```python
learn.lr_find(suggest_funcs=(valley, slide))
```

```python
learn.fine_tune(5, 0.01)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(5, 5))
```

```python
interp.plot_top_losses(5, nrows=3, figsize=(15,4))
```

```python
learn.export('140k-real-fake-face.pkl')
```
